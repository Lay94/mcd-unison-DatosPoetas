```python
import xml.etree.ElementTree as et 
import pandas as pd

import os  
import urllib.request 

import datetime 
```

```python
#Descargando archivo como xml
archivo_url = "https://github.com/mcd-unison/ing-caract/raw/main/ejemplos/integracion/ejemplos/wikipedia-poetas.xml"
archivo_nombre = "poetas.xml"
subdir = "./data/poetas/"

if not os.path.exists(subdir + archivo_nombre):
    if not os.path.exists(subdir):
        os.makedirs(subdir)
    urllib.request.urlretrieve(archivo_url, subdir + archivo_nombre)


poetas = et.parse(subdir + archivo_nombre)
```

## Visualizando estructura de los datos

```python
for node in poetas.iter():
    print(node.tag)
```

```python
for (i, poe) in enumerate(poetas.getroot()):
    if i > 0:
        print("Poeta {}:".format(i))
        for (j, prop) in enumerate(poe):
            if (j == 0):
                print("\t{}: {}".format(prop.tag, prop.text.strip()))
            elif (j == 3):
                for (k, rev) in enumerate(prop):                    
                    print("\t{}: {}".format(rev.tag, rev.text))
```

## Creando DataFrame

Se quiere contruir un dataframe, donde se encuentre almacenados los nombres de los poetas con sus urls 

```python
wiki_url = "https://es.wikipedia.org/wiki/"

poets_name = []
poets_url = []

for (i, poe) in enumerate(poetas.getroot()):
    if i > 0:
        for (j, prop) in enumerate(poe):
            if (j == 0):
                poets_name.append(format(prop.text.strip()))
                poets_url.append(wiki_url + format(prop.text.replace(" ", "_")))
df_poets = pd.DataFrame([poets_name, poets_url]).T
df_poets.columns = ['Poetas', 'URL']

df_poets
```

## Guardando el Dataframe como archivo tipo parquet

```python
df_poets.to_parquet('poets.parquet')
```
